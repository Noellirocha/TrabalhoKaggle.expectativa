* Gráficos de dispersão com linha de ajuste

```python
# Incidencia de HIV x expectativa de vida

import seaborn as sns            # Importação de bibliotecas
import scipy as sp               # Importação de bibliotecas
import matplotlib.pyplot as plt  # Importação de bibliotecas

plt.figure(figsize=(12, 8))                                                           # Tamanho ajustado para um gráfico

g = sns.lmplot(x='Incidents_HIV', y='Life_expectancy', data=data)                     # Criando o gráfico de dispersão com a linha de ajuste

# Função para adicionar a correlação no gráfico
def annotate(data, **kws):
    r, p = sp.stats.pearsonr(data['Incidents_HIV'], data['Life_expectancy'])
    ax = plt.gca()

    ax.text(0.05, 0.95, f'r = {r:.2f}, p = {p:.2g}', transform=ax.transAxes,          # Ajuste da posição do texto no gráfico
            fontsize=14, verticalalignment='top', horizontalalignment='left',
            bbox=dict(facecolor='white', edgecolor='none', boxstyle='round,pad=0.5'))


g.map_dataframe(annotate)              # Aplicando anotação


plt.xlabel('Incidentes de HIV', fontsize=14)                                          # Melhorando a apresentação do gráfico
plt.ylabel('Expectativa de Vida', fontsize=14)                                        # Melhorando a apresentação do gráfico

plt.show() # Exibindo o gráfico

# Interpretação: Podemos notar que cada número de novas infecções por HIV por 1.000 habitantes não infectados, reduz em média 0,55 na expectativa de vida.
```

* Gráficos de dispersão com linha de ajuste





```python
# Consumo de alcool X expectativa de vida

import seaborn as sns            # Importação de bibliotecas
import scipy as sp               # Importação de bibliotecas
import matplotlib.pyplot as plt  # Importação de bibliotecas

plt.figure(figsize=(12, 8))                                                             # Tamanho ajustado para um gráfico

g = sns.lmplot(x='Alcohol_consumption', y='Life_expectancy', data=data)                 # Criando o gráfico de dispersão com a linha de ajuste

# Função para adicionar a correlação no gráfico
def annotate(data, **kws):
    r, p = sp.stats.pearsonr(data['Alcohol_consumption'], data['Life_expectancy'])
    ax = plt.gca()

    ax.text(0.05, 0.95, f'r = {r:.2f}, p = {p:.2g}', transform=ax.transAxes,          # Ajuste da posição do texto no gráfico
            fontsize=14, verticalalignment='top', horizontalalignment='left',
            bbox=dict(facecolor='white', edgecolor='none', boxstyle='round,pad=0.5'))


g.map_dataframe(annotate)                                                             # Aplicando a anotação


plt.xlabel('Alcohol_consumption', fontsize=14)                                        # Melhorando a apresentação do gráfico
plt.ylabel('Expectativa de Vida', fontsize=14)                                        # Melhorando a apresentação do gráfico


plt.show() # Exibindo o gráfico


# Interpretação: Podemos notar que r = 0.42 indica uma correlação positiva entre o consumo de álcool e a expectativa de vida,
# caso fosse negativo, indicaria uma correlação inversa, ou seja quanto maior o consumo de álcool, menor a expectativa de vida.
# Em resumo, quanto maior renda disponivel, maior o consumo.
```

```python
mortalidade_infantil = data.groupby('Region')['Infant_deaths'].mean().round() # Criando um novo objeto para receber mortalidade_infantil, e agrupando e tirando a média
print (mortalidade_infantil)

# Interpretação:  Média de mortalidade infantil por região, lembrando que o número de mortes infantis por 1.000 nascidos vivos.
```

```python
mortalidade_adulta = data.groupby('Region')['Adult_mortality'].mean().round() #Criando um novo objeto para receber mortalidade_adulta, e agrupando e tirando a média
print (mortalidade_adulta)

# Interpretação:  Média de mortalidade adulta por região, lembrando que a taxa de mortalidade de adultos por 1.000 habitantes.
```

```python
mortalidade_menor_5 = data.groupby('Region')['Under_five_deaths'].mean().round() #Criando um novo objeto para receber mortalidade_menor_5 , e agrupando e tirando a média
print (mortalidade_menor_5)

# Interpretação:  Média de mortalidade menores de 5 anos por região, lembrando que o número de mortes de crianças menores de cinco anos por 1.000 nascidos vivos.
```

```python
escolaridade = data.groupby('Region')['Schooling'].mean().round() #Criando um novo objeto para receber escolaridade, e agrupando e tirando a média
print (escolaridade)

# Interpretação: Média de anos de escolaridade por região.
```

# **OKRs** - É uma metodologia de gestão e metas que ajudam a inspirar a inovação.



```python
from sklearn.linear_model import LinearRegression # Importação de bibliotecas
import statsmodels.api as sm                      # Importação de bibliotecas
import numpy as np                                # Importação de bibliotecas

# Criando um modelo de regressão linear

x = data[['Adult_mortality', 'GDP_per_capita_log', 'Infant_deaths', 'Schooling', 'Alcohol_consumption', 'Hepatitis_B', 'Polio', 'Incidents_HIV', 'BMI']]
y = data['Life_expectancy']
x, y = np.array(x), np.array(y)

x = sm.add_constant(x)       # Converte os dados em arrays NumPy
model = sm.OLS(y, x).fit()
print(model.summary())     # Exibe um resumo estatístico do modelo ajustado

# Coeficientes das variáveis: Mostram como cada variável explicativa afeta a variável alvo.
# Valor p: Indica se a relação entre a variável explicativa e a variável alvo é estatisticamente significativa (onde, 𝑝<0.05).

# Interpretação:
# 1º O R-quared (R2) ajustado indica que 98% da expectativa de vida pode ser explicada pelo modelo e seu respectivo conjunto de dados.
# 2º Analisando a aderência das variáveis com  P-valor adequado, isto é menor que 0,05 , podemos desconsiderar para fins de análise as variáveis x4 (escolaridade) e x8 (HIV)
# 3º Entre as variáveis adequadas e que possuem relação negativa podemos destacar x1 (mortalidade adulta) que para cada uma morte adicional reduz em 0,04 anos na expectativa de vida,
#  assim como x6 (Hepatitite) e x9 (Indice  de Massa Corporal), ou seja, a incidência de doenças graves e obesidade reduz a expectativa de vida da população.
# 4º Entre as variáveis com relação positiva, destacamos a x2 (renda per capita) que, por ser em log, tem de ser interpretada como variação percentual, ou seja,
# para cada um 1 % de aumento de renda eleva em 0,3 anos de vida.
# 5º Contudo, a associação positiva do x5 (consumo de alcool) com a expectativa de vida pode ser explicada pelo ultimo gráfico
# (Gráficos de dispersão com linha de ajuste - Consumo de alcool vs expectativa de vida.)
# em que o consumo de alcool está associado primeiramente a renda disponível, problema esse denominado de endogeneidade,
# ou seja, um problema de relação entre as variáves explicativas, pela literatura.
```
